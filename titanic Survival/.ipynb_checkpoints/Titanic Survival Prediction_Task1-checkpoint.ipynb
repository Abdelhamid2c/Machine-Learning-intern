```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.heatmap(titanic.isnull(), cbar=True, yticklabels=False)
plt.title('Distribution of Null Values')
plt.show()
```

```python
# Remplir les valeurs manquantes dans la colonne 'Age' du jeu de donnÃ©es d'entraÃ®nement
# en utilisant la mÃ©diane de cette colonne, et appliquer le remplacement directement dans le DataFrame.

moy_age = titanic['Age'].median()
titanic['Age'].fillna(moy_age, inplace=True)

titanic['Fare'].fillna(titanic['Fare'].median(), inplace=True)
titanic.drop(columns=['Cabin','Ticket'], axis=1, inplace=True)

titanic['Embarked'].fillna('S', inplace=True)


titanic.isnull().sum()
```

```python
#titanic.sample(3)
titanic.Pclass.value_counts()
```

```python
survival_percentage = titanic['Survived'].sum() / len(titanic) * 100
survival_percentage = round(survival_percentage)
not_survival_percentage = round(100 - survival_percentage)

plt.figure(figsize=(4, 4))
plt.pie([survival_percentage, not_survival_percentage], labels=['Survived', 'Not Survived'], autopct='%1.4f%%', colors=['green', 'red'], startangle=90)

# Set the aspect ratio to be equal, creating a circle
plt.axis('equal')

# Add a title
plt.title('Survival Percentage')

# Show the pie chart
plt.show()
```

```python
sns.barplot(x='Pclass',y='Survived',data=titanic);
```

```python

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 4))

class1 = titanic[titanic['Pclass'] == 1]
class2 = titanic[titanic['Pclass'] == 2]
class3 = titanic[titanic['Pclass'] == 3]

ax = sns.histplot(class1,
                  x='Sex',
                  bins=30,
                  hue='Survived',
                  multiple='stack',
                  ax=axes[0],
                  palette='RdBu',
                  kde=False)

ax.legend( labels=['Not Survived', 'Survived'])
ax.set_title("Class 1")

ax = sns.histplot(class2,
                  x='Sex',
                  bins=30,
                  hue='Survived',
                  multiple='stack',
                  ax=axes[1],
                  palette='RdBu',
                  kde=False)

ax.legend(labels=['Not Survived', 'Survived'])
ax.set_title("Class 2")

ax = sns.histplot(class3,
                  x='Sex',
                  bins=30,
                  hue='Survived',
                  multiple='stack',
                  ax=axes[2],
                  palette='RdBu',
                  kde=False)

ax.legend(labels=['Not Survived', 'Survived'])
ax.set_title("Class3")


plt.show()
```

```python

correlation_matrix = titanic.drop(columns=['PassengerId','Name','Sex','Embarked'],axis=1).corr()

plt.figure(figsize=(7, 5))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)

plt.title('Matrice de CorrÃ©lation')

plt.show()
```
